**Cuartiles** ($Q_1, Q_2, Q_3$) : son medidas de ubicación que dividen el conjunto de datos en cuatro grupos con aproximadamente el 25% de los valores en cada uno.

+ $Q_1$ (*primer cuartil*): valor que separa 25% de los valores inferiores ordenados del 75% superior.

+ $Q_2$ (*segundo cuartil*): valor equivalente a la **mediana**, que separa 50 \% de los valores inferiores ordenados del 50% superior.

+ $Q_3$ (*tercer cuartil*): valor que separa 75% de los valores inferiores ordenados del 25% superior.

<font color=blue>**Paso 5**</font> Estimación, intervalo de confianza e inferencia

Estimar el **porcentaje** de alumnos con una altura mayor a 1.65 m

$$\hat{p} = \frac{X}{n} $$
donde $X$ representa el número de alumnos con altura mayor a 1.65 m y $n$ el total de alumnos encuestados

```r
n=nrow(mydata)
ph=nrow(mydata[mydata$Height>1.65,])/n
print('Porcentaje de alumnos con altura mayor a 1.65 m =')
print(ph)
```

Generar un **intervalo de confianza del 90%** ($\alpha=0.1$ *nivel de significancia*) para la estimación de esta proporción

$$  p \in \left ( p_{min}, p_{max}  \right ) $$
donde


$$ p_{min}  = \frac{\hat{p} + \frac{z_{\alpha/2}^2}{2n} - z_{\alpha/2} \sqrt{\frac{\hat{p}\hat{q}}{n} + \frac{z_{\alpha/2}^2}{4n^2}}}{1+ \frac{z_{\alpha/2}^2}{n}} $$
$$p_{max}  = \frac{\hat{p} + \frac{z_{\alpha/2}^2}{2n} + z_{\alpha/2} \sqrt{\frac{\hat{p}\hat{q}}{n} + \frac{z_{\alpha/2}^2}{4n^2}}}{1+ \frac{z_{\alpha/2}^2}{n}} $$
donde $z_{\alpha/2} >0$ representa el **valor crítico** tal que
$$ P\left (-z_{\alpha/2} < Z < z_{\alpha/2}, \; Z  \sim \mathcal{N}(0,1)  \right ) = 1-\alpha $$
y $\hat{q}=1-\hat{p}$.

```r
qh=1-ph
alpha=0.1
za2=qnorm(1-alpha/2)
d=1+za2*za2/n
pmin=(ph+(za2**2)/(2*n)-za2*sqrt(ph*qh/n + (0.5*za2/n)**2))/d
pmax=(ph+(za2**2)/(2*n)+za2*sqrt(ph*qh/n + (0.5*za2/n)**2))/d
phi=c(pmin,pmax)
print('El intervalo de confianza del 90% es')
print(phi)
```

Una **hipótesis estadística** es una afirmación acerca del valor de uno o varios parámetros de una población.

Una **prueba de hipótesis estadística** es un procedimiento para probar una hipótesis sobre una propiedad de una población en función de una muestra.

<font color=blue>Evaluar la **hipótesis**: *El porcentaje de alumnos con altura mayor a 1.65 m es mayor al 60% ($p_0=0.6$) con un nivel de significancia del 5%*  ($\alpha=0.05$).</font>

**Hipótesis nula** $H_0: \hat{p}=p_0$,

**Hipótesis alternativa** $H_a: \hat{p}>p_0$.

El estadístico de prueba
$$ z = \frac{\hat{p}-p_0}{\sqrt{p_0(1-p_0)/n}} $$
y se genera la condición de **rechazo** de la hipotesis nula $H_0$:
$$ z \geq  z_\alpha$$
recordando como se obtiene el **valor crítico** $z_{\alpha}$:
$$\textrm{P}({Z} \geq z_{\alpha}, \; Z \sim \mathcal{N}(0,1) ) = \alpha  $$

```r
alfa=0.05
po=0.6
z=(ph-po)/sqrt(po*(1-po)/n)
za=qnorm(alfa, lower.tail = FALSE)
print('La condición de rechazo de H_0 es')
print(z >= za)
```

También se puede evaluar la prueba de hipótesis en función del *p-valor* en función del valor del estadístico de prueba $z$:

$$\textit{p-valor}=\textrm{P}( X \geq z, \; X \sim \mathcal{N}(0,1))$$
Entonces se **rechaza** la hipótesis nula $H_0$ si se cumple
$$ \textit{p-valor} < \alpha $$

```r
pvalor=pnorm(z,lower.tail=FALSE)
print('La condición de rechazo de H_0 es')
print(pvalor < alfa)
```
